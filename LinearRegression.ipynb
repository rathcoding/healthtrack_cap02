```python
'''
UNOCHAPECÓ
Ciência da Computação
Estatística e Probabilidade Discreta
Rafael Rath

Resolução de Regressão Linear para um Polinomio de qualquer grau
'''
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
#=== DEFINIÇÃO DE FUNÇÕES ======================================================
def rlp (X, Y, grau):   # Regressão Linear Polinomial
    Ar = np.empty([0,len(X)])
    for i in range(grau, 0, -1):
        Ar = np.append(Ar, [np.power(X,i)], axis = 0)
    Ar = np.append(Ar, [np.ones(len(X))], axis = 0)
    coeficientes = np.dot(np.linalg.inv((np.dot(Ar, Ar.T))), (np.dot(Ar, Y)))
    plota(coeficientes, Ar, X, Y)
    return coeficientes

def plota(coeficientes, Ar, X, Y):
    plt.plot(X, Y, 'ro')
    plt.plot(X, np.dot(Ar.T,coeficientes))
    plt.show()
    return

#=== INICIO DO PROGRAMA ========================================================
if __name__ == "__main__":
    print("Questão 1, letra A. Gabarito: y = 0.2167x + 0.1750 e R² = 0.9571")
    X = [1., 2., 3., 4., 5., 6., 7., 8.]
    Y = [0.5, 0.6, 0.9, 0.8, 1.2, 1.5, 1.7, 2.0]
    print(rlp(X,Y,1))

    print("\nQuestão 1, letra B. Gabarito: y = 0.0155x² + 0.0774x + 0.4071 e R² = 0.9767")
    print(rlp(X,Y,2))

    print("\nQuestão 3, letra b. Gabarito: y = −1.2616x² + 8.1165x − 6.1858")
    X = [0.5, 0.75, 1., 1.5, 2., 2.5, 3.]
    Y = [-2.8, -0.6, 1., 3.2, 4.8, 6., 7.]
    print(rlp(X,Y,2))
```
